### Load data

```python
import pandas as pd
import numpy as np
import qnorm
import pickle
```

```python
cytokine_df = pd.read_csv('/preprocessing/cytokine_data.csv',index_col = 0)
```

```python
#After Underbar (_) means time points sample number
cytokine_df
```

```python
patient_info = dict()
patient_info['Patient1'] = pd.read_csv('/preprocessing/patient1_clinical.csv',index_col = 0)
patient_info['Patient2'] = pd.read_csv('/preprocessing/patient2_clinical.csv',index_col = 0)
```

### Preprocessing

```python
cytokine_df = np.log1p(cytokine_df)
```

```python
cytokine_df = qnorm.quantile_normalize(cytokine_df,axis = 0)
cytokine_df
```

### Make meta data 

```python
patient_list = list(set([i[:-2] for i in cytokine_df.index if 'Healthy' not in i]))
```

```python
sample_name = []
patient_name = []
severity = []
max_severity = []
```

```python
    for i in cytokine_df.index:
        if 'Healthy' not in i:
            p_name = i[:-2]
            sample_name.append(i)
            patient_name.append(p_name)
            tmp_df = patient_info[p_name].copy()
            tmp_max = np.nanmax(tmp_df.WHO)
            tmp_severity = tmp_df[tmp_df.cytokine_exist == i].WHO.tolist()[0]
            severity.append(tmp_severity)
            if tmp_max >= 6:
                max_severity.append('Max_S')
            else:
                max_severity.append('Max_M')
    patient_meta = pd.DataFrame({'Sample':sample_name,'Patient':patient_name,'Severity':severity,'Max_Severity':max_severity})
```

```python
patient_meta
```

### Save preprocessed data

```python
preproc_data = {}
preproc_data['cytokine_data'] = cytokine_df
preproc_data['metadata'] = patient_meta
preproc_data['clinical'] = patient_info
with open('/preprocessing/processed_data.pickle','wb') as f:
    pickle.dump(preproc_data,f)
```
